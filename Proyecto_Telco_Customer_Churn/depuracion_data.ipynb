```python
import pandas as pd
import numpy as np
```

```python
df_telco=pd.read_csv("data\WA_Fn-UseC_-Telco-Customer-Churn.csv")
df_telco
```

Validar Duplicados

```python
df_telco.duplicated().sum()
```

```python
df_telco.info()
```

Clasificacion Genero

```python
df_telco.loc[df_telco['gender'] == 'Male', 'genero_bi'] = 'Masculino'
df_telco.loc[df_telco['gender'] == 'Female', 'genero_bi'] = 'Femenino'
```

Clasificacion si es adulto mayor

```python
df_telco.loc[df_telco['SeniorCitizen'] == 1,'es_adultoMayor'] = 'Si'
df_telco.loc[df_telco['SeniorCitizen'] == 0, 'es_adultoMayor'] = 'No'
```

Clasificacion si es Pareja

```python
df_telco.loc[df_telco['Partner'] == "Yes",'pareja_bi'] = 'Si'
df_telco.loc[df_telco['Partner'] == "No", 'pareja_bi'] = 'No'
```

Clasificacion si es Dependiente

```python
df_telco.loc[df_telco['Dependents'] == "Yes",'dependiente_bi'] = 'Si'
df_telco.loc[df_telco['Dependents'] == "No", 'dependiente_bi'] = 'No'
```

Rango de cuanto tiempo tiene con el servicio

```python
condiciones_dorm = [
    df_telco['tenure'].between(1, 6),
    df_telco['tenure'].between(7, 12),
    df_telco['tenure'].between(13, 24),
    df_telco['tenure'].between(25, 36),
    df_telco['tenure'].between(37, 48),
    df_telco['tenure'].between(49, 60),
    df_telco['tenure'] > 60
]

seleccion_dorm = [
    '1-6',
    '7-12',
    '13-24',
    '25-36',
    '37-48',
    '49-60',
    '> 60'
]

df_telco['tiempo_servicio_bi'] = np.select(condiciones_dorm, seleccion_dorm, default='Sin Consumo')
```

Clasificacion de Multiples lineas

```python
condicion_lineas = [
    df_telco['MultipleLines']=="Yes",
    df_telco['MultipleLines']=="No",

]

seleccion_lineas = [
    "Si",
    "No"
]

df_telco['Multiple_lineas_bi'] = np.select(condicion_lineas, seleccion_lineas, default='Sin Servicio Telefonico')
```

Tipos de Servicios de Internet

```python
condicion_lineas_internet = [
    df_telco['InternetService']=="DSL",
    df_telco['InternetService']=="Fiber optic",

]

seleccion_lineas_internet = [
    "DSL",
    "Fibra Optica"
]

df_telco['InternetService_bi'] = np.select(condicion_lineas_internet, seleccion_lineas_internet, default='No tiene Servicio Internet')
```

Servicio telefonico

```python
df_telco['Servicio_Telefonico_bi']="Sin Servicio Internet"
df_telco.loc[df_telco['PhoneService'] == "Yes",'Servicio_Telefonico_bi'] = 'Si'
df_telco.loc[df_telco['PhoneService'] == "No", 'Servicio_Telefonico_bi'] = 'No'
```

Proteccion de Equipo

```python
df_telco['proteccion_equipo_bi']="Sin Servicio Internet"
df_telco.loc[df_telco['DeviceProtection'] == "Yes",'proteccion_equipo_bi'] = 'Si'
df_telco.loc[df_telco['DeviceProtection'] == "No", 'proteccion_equipo_bi'] = 'No'
```

Respaldo de info

```python
df_telco['Respaldo_info_bi']="Sin Servicio Internet"
df_telco.loc[df_telco['OnlineBackup'] == "Yes",'Respaldo_info_bi'] = 'Si'
df_telco.loc[df_telco['OnlineBackup'] == "No", 'Respaldo_info_bi'] = 'No'
```

soporte tecnico

```python
df_telco['Soporte_tecnico_bi']="Sin Servicio Internet"
df_telco.loc[df_telco['TechSupport'] == "Yes",'Soporte_tecnico_bi'] = 'Si'
df_telco.loc[df_telco['TechSupport'] == "No", 'Soporte_tecnico_bi'] = 'No'
```

Streaming general

```python
df_telco['Streaming_bi']="Sin Servicio Internet"
df_telco.loc[df_telco['StreamingTV'] == "Yes",'Streaming_bi'] = 'Si'
df_telco.loc[df_telco['StreamingTV'] == "No", 'Streaming_bi'] = 'No'
```

Streaming de pelicula

```python
df_telco['streaming_peliculas_bi']="Sin Servicio Internet"
df_telco.loc[df_telco['StreamingMovies'] == "Yes",'streaming_peliculas_bi'] = 'Si'
df_telco.loc[df_telco['StreamingMovies'] == "No", 'streaming_peliculas_bi'] = 'No'
```

Tiempo de contrato

```python
condicion_contrato = [
    df_telco['Contract']=="Month-to-month",
    df_telco['Contract']=="Two year",
    df_telco['Contract']=="One year",

]

seleccion_contrato = [
    "Mes a Mes",
    "Un Año",
    "Dos Años",
]

df_telco['contrato_bi'] = np.select(condicion_contrato, seleccion_contrato, default='Sin Servicio Telefonico')
```

Facturacion Electronica

```python
df_telco.loc[df_telco['PaperlessBilling'] == "Yes",'Faturacion_electronica_bi'] = 'Si'
df_telco.loc[df_telco['PaperlessBilling'] == "No", 'Faturacion_electronica_bi'] = 'No'
```

metodos de pago

```python
condicion_pago = [
    df_telco['PaymentMethod']=="Electronic check",
    df_telco['PaymentMethod']=="Mailed check",
    df_telco['PaymentMethod']=="Bank transfer (automatic)",
    df_telco['PaymentMethod']=="Credit card (automatic)",
]

seleccion_pago = [
    "Cheque Electronico",
    "Cheque por Correo",
    "Transferencia Bancaria",
    "Tarjeta de Credito",
]

df_telco['FormaPago_bi'] = np.select(condicion_pago, seleccion_pago, default='Sin Servicio Telefonico')
```

si mi cliente se queda en el plan de contrato

```python
df_telco.loc[df_telco['Churn'] == "Yes",'Churn_bi'] = 'Clientes No Retenidos'
df_telco.loc[df_telco['Churn'] == "No", 'Churn_bi'] = 'Clientes Retenidos'
```

Las columnas que no necesito

```python
df_telco = df_telco.drop(columns=["gender","SeniorCitizen","Partner",
                                   "Dependents","tenure","PhoneService","MultipleLines",
                                   "OnlineSecurity","OnlineBackup","Churn","PaymentMethod",
                                   "PaperlessBilling","StreamingTV","TechSupport","DeviceProtection","Contract",
                                   "StreamingMovies","InternetService"])
```
